```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# wczytaj 'Mall_Customers.csv'
df = pd.read_csv('/usr/share/SI/Mall_Customers.csv')
```

```python
#1. usuń niepotrzebne kolumny z df
#2. zamień dane tekstowe na liczbowe
```

```python
from sklearn.cluster import KMeans

X_train = df

kmeans = KMeans(n_clusters=6)
model = kmeans.fit(X_train)
labels = model.predict(X_train)


fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.scatter(df["Age"], df["Annual Income (k$)"], df["Spending Score (1-100)"], c=labels)

ax.set_xlabel('Age')
ax.set_ylabel('Annual Income')
ax.set_zlabel('Spending Score')
```
